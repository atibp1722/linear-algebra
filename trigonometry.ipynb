#### Trigonometric functions plot

```python
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-2*np.pi,2*np.pi,100)
sin_v=np.sin(x)
cos_v=np.cos(x)
tan_v=np.tan(x)

plt.figure(figsize=(8,5))
plt.plot(x,sin_v,label='sin(x)',color='red')
plt.plot(x,cos_v,label='cos(x)',color='blue')
plt.plot(x,tan_v,label='tan(x)',color='green',alpha=0.7)
plt.title('Basic Trigonometric Plot')
plt.xlabel('Radians')
plt.ylabel('Function value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x=np.linspace(-1*np.pi,1*np.pi,100)
sec_v=1/np.cos(x)
cosec_v=1/np.sin(x)
cot_v=1/np.tan(x)

plt.figure(figsize=(8,5))
plt.plot(x,sec_v,label='sec(x)',color='green')
plt.plot(x,cosec_v,label='cosec(x)',color='red')
plt.plot(x,cot_v,label='cot(x)',color='blue',alpha=0.7)
plt.title('Basic Trigonometric Plot')
plt.xlabel('Radians')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x=np.linspace(0,5*np.pi,1000)
a=np.sin(x)
b=5*np.sin(x-np.pi/5)

plt.figure(figsize=(8,5))
plt.plot(x,a,label='sin(x)',linestyle='--')
plt.plot(x,b,label='5*sin(x-π/5)',color='green')
plt.title('Shift Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
sin_v=np.linspace(-1,1,100)
cos_v=np.linspace(-1,1,100)
tan_v=np.linspace(-5,5,1000)

arcsin_v=np.arcsin(sin_v)
arccos_v=np.arccos(cos_v)
arctan_v=np.arctan(tan_v)

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.plot(sin_v,arcsin_v,'green',linewidth=2,label='arcsin(v)')
plt.title('Inverted Sin Function')
plt.xlabel('x')
plt.ylabel('arcsin(v)')
plt.grid(True,alpha=0.3)
plt.legend()

plt.subplot(1,3,2)
plt.plot(cos_v,arccos_v,'blue',linewidth=2,label='arccos(v)')
plt.title('Inverted Cosine Function')
plt.xlabel('x')
plt.ylabel('arccos(v)')
plt.grid(True,alpha=0.3)
plt.legend()

plt.subplot(1,3,3)
plt.plot(tan_v,arctan_v,'red',linewidth=2,label='arctan(v)')
plt.title('Inverse Tan Function')
plt.xlabel('x')
plt.ylabel('arctan(x)')
plt.grid(True,alpha=0.3)
plt.legend()

plt.tight_layout()
plt.show()
```
